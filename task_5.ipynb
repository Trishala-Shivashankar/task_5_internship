```python
import pandas as pd
file_path = "TITANIC DATASET.csv" 
df = pd.read_csv(file_path)
print("Initial Data Snapshot:")
display(df.head())
```

```python
df.info()
display(df.isnull().sum())
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop('Cabin', axis=1, inplace=True)

display(df.isnull().sum())
df.info()
```

```python
print("Descriptive Statistics:\n", df.describe())
print("\nSurvived Value Counts:\n", df['Survived'].value_counts())
print("\nPclass Value Counts:\n", df['Pclass'].value_counts())
print("\nSex Value Counts:\n", df['Sex'].value_counts())
print("\nEmbarked Value Counts:\n", df['Embarked'].value_counts())
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived')
plt.suptitle("Pairplot of Numerical Features by Survival", y=1.02)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
df['Age'].hist(bins=30, edgecolor='black')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Fare'].hist(bins=30, edgecolor='black')
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title("Boxplot of Age by Pclass")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title("Boxplot of Fare by Survival")
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Fare vs Age Colored by Survival")
plt.show()
```
